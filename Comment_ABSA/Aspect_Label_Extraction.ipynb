## Explanation of 3rd Phase 

In this step, we aim to extract the aspect features from the pre-processed data. For this phase I will employ the **Comment_Sentence_List** and **Advantage_Sentences** and **Disadvantage_Sentences**. 

```python
for i in range(len(data)):
    comment_sentences_list = []
    comment_sentences_list = data.iloc[i]["Comment_Sentence_List"]
    for j in range(len(comment_sentences_list)):
        comment_sentences_list[j] = comment_sentences_list[j].split()
        
        
```

```python
edited_data = ast.literal_eval(data.iloc[0]["Advantage_Sentences"])
```

```python
edited_data[0][0]
```

```python
temp_adv_list = []
for adv_el in range(len(edited_data)):

    temp_sentence_list = []
    for adv_ph_el in range(len(edited_data[adv_el])):
        try:
            word = re.search(r'\(([^\,]+)', edited_data[adv_el][adv_ph_el]).group(1)
            
        except Exception as e: 
            print("The Error is : ", e)
            print("Error on record: {0}, element {1} function {2} ".format(adv_el,adv_ph_el,"Word Extraction"))
            word = ''
                        
        try:
            PoS = re.search(r'\,([^\)]+)', edited_data[adv_el][adv_ph_el]).group(1)
        except Exception as e: 
            print("The Error is : ", e)
            print("Error on record: {0}, element {1} function {2} ".format(adv_el,adv_ph_el,"PoS Extraction"))
            PoS = ''
                        
        temp = tuple([word,PoS])
        temp_sentence_list.append(temp)
    temp_adv_list.append(temp_sentence_list)
```

```python
print(re.search(r'\(([^\,]+)', edited_data[0][0]).group(1))   # Outputs the word 
print(re.search(r'\,([^\)]+)', edited_data[0][0]).group(1))   # Outputs the PoS
```
